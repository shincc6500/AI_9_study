```python
import pandas as pd
import numpy as np
```

```python
# 1. Python 라이브러리 함수를 사용하여 엑셀 파일을 불러오고, DataFrame을 출력해주세요.
df_excel = pd.read_excel("관서별 5대범죄 발생 및 검거.xlsx") #excel 불러오기

df_excel #dataframe 출력.
```

```python
#2. 각 경찰서(관서명)를 해당 구 이름으로 매핑하여 '구별'이라는 새로운 column을 생성하고, DataFrame을 출력해주세요.
tmp = {'서대문서': '서대문구', '수서서': '강남구', '강서서': '강서구', '서초서': '서초구',
'서부서': '은평구', '중부서': '중구', '종로서': '종로구', '남대문서': '중구',
'혜화서': '종로구', '용산서': '용산구', '성북서': '성북구', '동대문서': '동대문구',
'마포서': '마포구', '영등포서': '영등포구', '성동서': '성동구', '동작서': '동작구',
'광진서': '광진구', '강북서': '강북구', '금천서': '금천구', '중랑서': '중랑구',
'강남서': '강남구', '관악서': '관악구', '강동서': '강동구', '종암서': '성북구', 
'구로서': '구로구', '양천서': '양천구', '송파서': '송파구', '노원서': '노원구', 
'방배서': '서초구', '은평서': '은평구', '도봉서': '도봉구'} #구별 컬럼 생성하기 위한 딕셔너리 데이터. 

#구별 컬럼 생성하기 위한 병합 데이터 프레임 생성 및 가공.
tmp = pd.DataFrame(tmp, index= ['구별']) #입력받은 딕셔너리를 DataFrame로 변환, 컬럼 명을 구별로 입력.
tmp = tmp.T #데이터 프레임을 가로에서 세로로 변환
tmp  = tmp.reset_index(drop=False) #index를 초기화하고 기존 index를 컬럼으로 이동
tmp = tmp.rename(columns={'index': '관서명'}) #컬럼 명을 재정의 

df2 = pd.merge(df_excel,tmp, on= '관서명', how = 'left') #df_excel에 관서명에 따라 구별 컬럼 추가. 문제 번호에 따라 df에 번호 부여
df2 = df2.fillna('구없음') #nan 값 대신에 구없음 입력. 
df2
```

```python
# pivot_table 을 사용하여 관서별 데이터를 구별 데이터로 변경하고, 같은 구의 경우에는 sum을 적용하여 더해주세요. (index : 관서명 -> 구별)
df3 = pd.pivot_table(df2,
                     index = '구별',
                     aggfunc = "sum") #구별 컬럼을 index로 지정. 문제마다 데이터 프레임을 구분하여 각 문제마다 독립적으로 사용. 
df3 = df3.drop('관서명', axis =1) #기존 관서명 컬럼 제거. 
df3 = df3.groupby('구별').sum() #구별로 묶어서 같은 컬럼 값끼리 합을 출력.
df3
```

```python
# 4. 구 없음  행은 drop 을 활용하여 삭제해주세요.
df4= df3.copy() #문제마다 데이터 프레임을 구분하여 각 문제마다 독립적으로 사용. 
df4 = df4.drop('구없음', axis =0) #'구없음' 행을 제거. 
df4
```

```python
# 5. 각 범죄 별로 검거율을 계산하고, 각 검거율 데이터 column을 DataFrame에 추가해주세요.
df5 = df4.copy() #문제마다 데이터 프레임을 구분하여 각 문제마다 독립적으로 사용. 
crime_list = ['살인','강도','강간','절도','폭력'] #반복문에서 사용하기 위한 범죄 종류.

#검거율 컬럼 생성.
i = '소계' #반복 줄이기 위한 호출변수
df5['검거율'] = df5[f'{i}(검거)']/df5[f'{i}(발생)']*100
#반복문을 이용한 범죄에 따른 검거율 컬럼 생성
for i in crime_list:
    df5[f'{i}검거율'] = df5[f'{i}(검거)']/df5[f'{i}(발생)']

df5
```

```python
# 6.필요없는 column을 del 을 사용하여 삭제해주세요.
df6 = df5.copy() #문제마다 데이터 프레임을 구분하여 각 문제마다 독립적으로 사용. 
delete_list = ['강간(검거)','강도(검거)','살인(검거)','절도(검거)','폭력(검거)','소계(발생)','소계(검거)'] #제거해야 할 컬럼 리스트\
#반복문을 통해 delete_list안의 컬럼을 제거.
for i in delete_list:
    del df6[i]
df6
```

```python
# 7. DataFrame의 컬럼명을 rename 을 사용하여 변경해주세요.
df7 = df6.copy() #문제마다 데이터 프레임을 구분하여 각 문제마다 독립적으로 사용. 
df7 = df7.rename(columns={'강간(발생)':'강간','강도(발생)':'강도','살인(발생)':'살인','절도(발생)':'절도','폭력(발생)':'폭력'}) #컬럼 명 변경
df7
```

```python
#extra 1 Python 라이브러리 함수를 사용하여 인구 데이터(pop_kor.csv) 파일을 불러오고, DataFrame을 출력해주세요.
df_csv = pd.read_csv("pop_kor.csv",index_col= '구별') #csv 불러오기
df_csv
```

```python
#extra 2 join 을 사용하여 Quiz에서 수행한 DataFrame과 인구 데이터 DataFrame을 merge하고, DataFrame을 출력해주세요.
extra_df2 = df7.join(df_csv) #df7에 df_csv를 index를 기준으로 결합. 
extra_df2
```

```python
extra_df3 = extra_df2.copy() #문제마다 데이터 프레임을 구분하여 각 문제마다 독립적으로 사용. 
extra_df3 = extra_df3.sort_values('검거율',ascending= True) #검거율 기준으로 오름차순 정렬
extra_df3
```
